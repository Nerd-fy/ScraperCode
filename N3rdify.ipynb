```python
lista_manga=[]
import time
elenco = wd.find_elements_by_css_selector('#row-elenco-opere > div > div > div.caption.text-center > h5')
print(len(elenco))
for e in elenco:
        manga = e.text
        link = e.find_elements_by_css_selector('a')[0].get_attribute('href')
        lista_manga.append({'manga': manga,
                            'link': link})
while len(elenco) == 60:
    time.sleep(2)
    page = wd.find_elements_by_css_selector('#pagination-opera > li.page-item.next > a')[0]
    page.click()
    time.sleep(2)
    elenco = wd.find_elements_by_css_selector('#row-elenco-opere > div > div > div.caption.text-center > h5')
    print(len(elenco))
    for e in elenco:
        manga = e.text
        link = e.find_elements_by_css_selector('a')[0].get_attribute('href')
        lista_manga.append({'manga': manga,
                            'link': link})
```

```python
import pprint
print(len(lista_manga))
```

```python
import pandas as pd
ds_lista_manga = pd.DataFrame(lista_manga)
ds_lista_manga.head()
wd.quit()
```

```python
ds_lista_manga.to_csv('lista_manga.csv')
```

```python
for m in tqdm(ds_lista_manga['link']):
    time.sleep(4)
    wd = webdriver.Chrome('chromedriver',options=chrome_options)
    wd.get(m)
    try:
      skip = wd.find_elements_by_css_selector('div:nth-child(1) > div > a')[0]
      skip.click()
    except Exception as e:
      print(e)
      continue
    campi = wd.find_elements_by_css_selector(".dl-horizontal > dt")
    valori = wd.find_elements_by_css_selector(".dl-horizontal > dd")
    manga_details = []
    for _ in range(len(campi)):
       manga_details.append("".join([campi[_].text,":",valori[_].text]))
    trama = wd.find_element_by_css_selector("#trama-div")
    manga_details.insert(-1,trama)
    wd.close()
print(len(manga_details))
```

```python
ds_manga_dettagli = pd.DataFrame(manga_details)
ds_manga_dettagli.head()
```
